```python
import os
os.chdir("/Users/jazminvaleriano/Library/Mobile Documents/com~apple~CloudDocs/03 UNIFR MS/00. SP25/00.MASTER_THESIS/FINAL_CHAPTERS/10_swONT_viral_metagenomics")

import pandas as pd

results_dir="04_positive_results"
metadata="metadata.csv"

dfs= [] # List to store all dfs

# Loop through all TSV files in the directory
for filename in os.listdir(results_dir):
    if filename.endswith(".tsv"):
        file_path = os.path.join(results_dir, filename)
        sample_name = os.path.basename(filename).split("_")[0]
        
        # Read the TSV file
        df = pd.read_csv(file_path, sep="\t")
        
        # Add the Sample column
        df['Sample'] = sample_name
        
        # Append to the list
        dfs.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# Reordering a little 
first_cols = ['Sample', 'Taxon', 'TaxonName', 'MappedReads']

# Get the rest of the columns, excluding the ones already in `first_cols`
remaining_cols = [col for col in combined_df.columns if col not in first_cols]
combined_df = combined_df[first_cols + remaining_cols]

# Load metadata
metadata = pd.read_csv("metadata.csv")

# Merge metadata into combined_df based on 'Sample'
combined_df = combined_df.merge(metadata, on="Sample", how="left")

combined_df.to_csv(f"{results_dir}/combined_report.tsv", sep="\t", index=False)
combined_df


```
