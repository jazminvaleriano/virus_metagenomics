```python
import os
os.chdir("/Users/jazminvaleriano/Library/Mobile Documents/com~apple~CloudDocs/03 UNIFR MS/00. SP25/00.MASTER_THESIS/FINAL_CHAPTERS")
print(os.getcwd())
import pandas as pd
import plotly.graph_objects as go

# --- File paths ---
paths = [
    "03_Bacterial_mNGS_validation_LONG/100K_ZM Long (length-adjusted)/100K_ZM Long (length-adjusted)-sp-chi-sqrd-0.001-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/100K_ZM Long (length-adjusted)/100K_ZM Long (length-adjusted)-sp-chi-sqrd-0.1-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/100K_ZM Long (length-adjusted)/100K_ZM Long (length-adjusted)-sp-chi-sqrd-1-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/20K_ZM Long (length-adjusted)/20K_ZM Long (length-adjusted)-sp-chi-sqrd-0.001-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/20K_ZM Long (length-adjusted)/20K_ZM Long (length-adjusted)-sp-chi-sqrd-0.1-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/20K_ZM Long (length-adjusted)/20K_ZM Long (length-adjusted)-sp-chi-sqrd-1-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/5K_ZM Long (length-adjusted)/5K_ZM Long (length-adjusted)-sp-chi-sqrd-0.001-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/5K_ZM Long (length-adjusted)/5K_ZM Long (length-adjusted)-sp-chi-sqrd-0.1-perc.csv",
    "03_Bacterial_mNGS_validation_LONG/5K_ZM Long (length-adjusted)/5K_ZM Long (length-adjusted)-sp-chi-sqrd-1-perc.csv"
]

for f in paths:
    if not os.path.exists(f):
        print("Missing file:", f)


# --- Load and concatenate all data ---
df_all = pd.concat([pd.read_csv(f) for f in paths], ignore_index=True)

# Extract size of ds
df_all['Size'] = df_all['dataset'].str.extract(r'^(5K|20K|100K)')


# Drop incomplete rows (e.g., MEGAN missing)
df_all = df_all.dropna(subset=['L1_distance'])

df_all
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for threshold = 1%
df_1perc = df_all[df_all['threshold'] == 1.0].copy()

# Prepare plotting variables
df_1perc['Classifier'] = pd.Categorical(df_1perc['method'], categories=[
    "KRAKEN", "CENTRIFUGER", "CENTRIFUGE", "KAIJU", "DIAMOND", "MEGAN"
], ordered=True)

df_1perc['Size'] = pd.Categorical(df_1perc['Size'], categories=["5K", "20K", "100K"], ordered=True)

pal_class = sns.color_palette(['#1b9e77', '#d95f02', '#7570b3', '#e7298a', '#66a61e', '#e6ab02'])

# Plot
sns.set_theme(style="whitegrid", font_scale=1)
plt.figure(figsize=(8, 4))
g = sns.swarmplot(
    data=df_1perc,
    x="L1_distance",
    y="Size",
    hue="Classifier",
    palette=pal_class,
    size=10,
    dodge=False
)

plt.xlabel("L1 Distance")
plt.ylabel("Dataset Size")
plt.title("Species L1 Distance by Classifier")
plt.legend(title="Classifier", bbox_to_anchor=(1.02, 0.5), loc="center left")
plt.tight_layout()
plt.savefig("Zymo_validation/distances_plot_1perc.svg", dpi=300, bbox_inches='tight')
plt.show()
```
